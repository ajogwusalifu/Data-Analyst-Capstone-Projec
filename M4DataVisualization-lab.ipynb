### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

QUERY = """
SELECT ConvertedComp, Age, WorkWeekHrs, CodeRevHrs FROM {}
""".format(table_name)

df = pd.read_sql_query(QUERY, conn)

# Filter data for the age group 30 to 35
filtered_df = df[(df['Age'] >= 30) & (df['Age'] <= 35)]

# Group by Age and calculate median for WorkWeekHrs and CodeRevHrs
median_values = filtered_df.groupby('Age')[['WorkWeekHrs', 'CodeRevHrs']].median()

# Create a stacked bar chart
median_values.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30 to 35')
plt.xlabel('Age')
plt.ylabel('Median Hours')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
df = pd.read_sql_query(QUERY, conn)

import pandas as pd
import matplotlib.pyplot as plt

QUERY = """
SELECT ConvertedComp, Age FROM {}
""".format(table_name)

# Example:
df = pd.read_sql_query(QUERY, conn)

# Filter data for ages between 45 and 60
filtered_data = df[(df['Age'] >= 45) & (df['Age'] <= 60)]

# Group data by age and calculate the median ConvertedComp
median_comp_by_age = filtered_data.groupby('Age')['ConvertedComp'].median()

# Create a line plot
plt.figure(figsize=(10, 6))
median_comp_by_age.plot(marker='o', linestyle='-', color='purple')
plt.title('Median ConvertedComp for Ages 45 to 60')
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.grid(True)
plt.show()

```
